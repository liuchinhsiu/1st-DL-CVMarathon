```python
names = ['b1', 'b2','b3']
data=dict(df[['b1','b2','b3']].transpose())

def bar_percent(data, category_names):
    # dictionary 取 key 為 labels
    labels = list(data.keys())
    # dictionary 取 values 為 data
    data = np.array(list(data.values()))   
    #print(data)
    # 各筆資料總合
    data_cum = data.cumsum(axis=1)
    # 決定類別的顏色
    category_colors = plt.get_cmap('RdYlGn')(np.linspace(0.15, 0.85, data.shape[1])) #設定選擇色條使用位置的參數
    # 圖區大小
    fig, ax = plt.subplots(figsize=(10, 5)) 
    # y 刻度
    ax.set_yticks(range(len(labels)))
    # y 刻度標籤
    ax.set_yticklabels(range(len(labels)))
    # y 由小到大
    ax.invert_yaxis()
    # x 不出現刻度
    ax.xaxis.set_visible(False)
    # 設定 x 範圍
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        # 寬度
        widths = data[:, i]
        # 啟始處，總合減去寬度
        starts = data_cum[:, i] - widths
        # 水平的 bar 繪圖
        ax.barh(labels, widths, left=starts, height=0.7,label=colname, color=color)
        # bar 中點位置
        xcenters = starts + widths / 2
        # 讀取顏色資訊
        r, g, b, _ = color
        # 決定文字顏色，有兩種，白色與灰色
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        # 加入文字
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',color=text_color)
    # 增加圖例        
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')
    
# 啟動函數
bar_percent(data, names)
```

## np.meshgrid(x,y)

```python
#由x y 決定z
def cal_z(x, y):
    cal = np.cos(y/10)*(x*y)**.5-x*0.3
    return cal

#製作一個網格，定義 x, y
x = np.linspace(1, 10, 100)
y = np.linspace(1, 10, 100)
#cal_z(x, y)

#由x, y構成網格
x, y = np.meshgrid(x, y)

#計算每一點網格的z值
z = cal_z(x, y)

#散佈圖，顏色與大小設定
matplotlib.rcdefaults()
fig, ax = plt.subplots(figsize=(8,8))

ax.set_title("scatter")

# 散佈圖，顏色愈淺，數字愈大
ax.scatter(x, y,c=z)

cmap=matplotlib.cm.Purples_r         # 紅藍反向
p = ax.pcolor(x, y, z, cmap=cmap, vmin=z.min(), vmax=z.max())

cb = fig.colorbar(p, ax=ax)
```

```python
# 中文設定
from matplotlib import rcParams
# 中文設定
rcParams["font.sans-serif"] = ["kaiu"] 
rcParams["font.family"] ="DFKai-sb"

# ggplot繪圖風格
plt.style.use('ggplot')

# 兩個個案比較，可多個比較，五個構面 *******
values = [3.2,2.1,3.5,2.8,3.2]
values2 = [4.9,4.1,4.5,5,4.1]
feature = ['構面一','構面二','構面三','構面四','構面五']

# 構面個數
N = len(values)

# 角度設定
angles=np.linspace(0, 2*np.pi, N, endpoint=False) #一個完整圓的弧度為2pi

# 使圖形封閉
values=np.concatenate((values,[values[0]])) #Join a sequence of arrays along an existing axis.
values2=np.concatenate((values2,[values2[0]]))
angles=np.concatenate((angles,[angles[0]]))

# 繪圖，設定圖面大小
fig=plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, polar=True)

# 第一條折線圖
ax.plot(angles, values, 'o-', linewidth=2, label = '個案一')
# 顏色
ax.fill(angles, values, alpha=0.25)


# 第二條折線圖
ax.plot(angles, values2, 'o-', linewidth=2, label = '個案二')
# 顏色
ax.fill(angles, values2, alpha=0.25)

# 標籤
ax.set_thetagrids(angles * 180/np.pi, feature) #180/pi轉化成度數



# 範圍
ax.set_ylim(0,5)

# 標題
plt.title('個案一與個案二在五個構面上的比較')

# 網格線
ax.grid(True)
# 圖例
plt.legend(loc = 'best')

# 顯示圖形
plt.show()
```
