# 中文編碼
* 中文編碼有別，一般為 utf-8，在台灣有時為 big5 
* nodepad++ 可進行中文編碼轉換


```python
# 依年代重新排序
doc_list=['蔣中正第一任.txt', '蔣中正第二任.txt','蔣中正第三任.txt','蔣中正第四任.txt','蔣中正第五任.txt','蔣經國第六任.txt', 
          '蔣經國第七任.txt','李登輝第九任.txt', '陳水扁第十任.txt', '陳水扁第十一任.txt' , 
          '馬英九第十二任.txt','馬英九第十三任.txt','蔡英文第十四任.txt']
#texts=''
docs=[]
i=0
# 讀取各篇文章
for f in doc_list: 
        i+=1
        # 完整檔名
        file_name=path +'/corpus/president/'+f
        # 讀取檔案，注意格式為 'utf8'，若為 big5 無毋須 encoding。
        # nodepad++ 可進行中文編碼轉換
        content = open(file_name,'r',encoding='utf8').read()  
        # 讀到的文字檔加入 list, docs
        docs.append(content)
        
        # 每處理完10篇文章，報告進度。******    
        if i%10==0:
            print (i, 'done')
        elif i==len(doc_list):
            print (i, 'finished')
```

```python
doc_list=['蔣中正第一任.txt', '蔣中正第二任.txt','蔣中正第三任.txt','蔣中正第四任.txt','蔣中正第五任.txt','蔣經國第六任.txt', 
          '蔣經國第七任.txt','李登輝第九任.txt', '陳水扁第十任.txt', '陳水扁第十一任.txt' , 
          '馬英九第十二任.txt','馬英九第十三任.txt','蔡英文第十四任.txt']

docs = []
i = 0
for f in doc_list:
    i+=1
    file_name = path + '/corpus/president/'+f
    content = open(file_name, 'r', encoding = 'utf8').read()
    docs.append(content)
    
    if i %10 == 0:
        print(i, 'done')
    elif i == len(doc_list):
        print(i, 'finished')
```

```python
# 讀取 station.txt
# 注意檔案編碼方式為 utf-8
coo = open(path+'data/station.txt', 'r',encoding='utf-8').read() #r 代表讀取模式
coo
```

```python
# 依換行符號切成 list
coo_line=coo.split('\n')
coo_line
```

```python
coo_line = coo.split('\n')
coo_line
```

### 座標資訊，文字檔，改成 dataframe 

```python
import pandas as pd
st_ind=[]
position=[]
city=[]
xcoo=[]
ycoo=[]
# readlines() 依行序讀入資料
for line in coo_line:
    # 依空格分解成list
    da=line.split()
    # 跳過空白行
    if len(da)>0:
        st_ind.append(da[0])   
        position.append(da[1])
        city.append(da[2])
        xcoo.append(da[3])
        ycoo.append(da[4])
data = list(zip(st_ind,position, city, xcoo, ycoo))
# 把第一列標題去掉
data=data[1:]
# DataFrame 加入欄位名稱
df = pd.DataFrame(data, columns=['st_ind', 'position', 'city', 'xcoo', 'ycoo'])
df.head()
```
