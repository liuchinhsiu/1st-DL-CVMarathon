```python
#####################################################################
import pandas as pd
path='d:/My python/'
df=pd.read_csv(path+"_PTT/Gossiping_ran2019_5000.csv", index_col=None)
df
```

```python
# 刪除無值的紀錄
df.text.isnull().value_counts()
df.text=df.text.fillna(999)
delete=df[df.text==999].index.tolist()
print (len(delete))
df=df.drop(delete)
df.index=range(len(df))
```

```python
# 刪除非字串
delete=[]
for n in range(len(df)):
    try:
        if (type(df.text[n])!= str) or (len(df.text)<1):
            delete.append(n)
    except:
        delete.append(n)
print (len(delete))     
df=df.drop(delete)
df.index=range(len(df))
```

```python
# 分詞前先去除特殊符號
import re
text=[]
for t in df.text:
    # 特殊符號可以自己增補
    t = re.sub(r"[\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]+", "",t)
    t = re.sub(r'[-，+、@,？⊂・ω{}●►❖★！!：\\/(=)…（）『』%《》$;；」:?=<>"／.&#"「_【】]', "",t)    
    # 去除數字
    t = re.sub(r"[0123456789]","",t)    
    # 去除英文  
    t = re.sub(r"[a-zA-Z]","",t)      
    text.append(t)
               
df.text=text              
```

# 結巴分詞

```python
import jieba
import jieba.analyse
from optparse import OptionParser
import re

# 文章內容轉存成 list
doc_list=list(df.text)
    
# 設定斷詞字典，未指定則為 default 的字典
jieba.set_dictionary(path + "corpus/dict.txt.big.txt")    
   
# 增加新詞，字典可以隨時叫出修改， 增添新的詞彙。 
jieba.load_userdict(path + "corpus/userdict.txt")   

# 停用字設定，請比較其差異 ######
jieba.del_word('政權和平移轉')
jieba.del_word('自由人')

texts=[]
i=0
# 讀取各篇文章
for d in doc_list: 
    i+=1        
    # 結巴分詞，引數為 string。
    if type(d)!=float:
        if len(d)>0:
            # 異體字轉換
            d=re.sub("台","臺",d)
            d=re.sub("\n","",d)
            d=re.sub("○","０",d)
            d=re.sub(" ","",d)
            # 刪除中文的空格符號，這些看起來像是亂碼先刪除 
            d=re.sub("\ufeff","",d)
            d=re.sub("\u3000","",d)          
            
            result = jieba.cut(d)       
            # 讀取中文斷詞結果，存為 string，中間以空白隔開
            seg=''
            for word in result:
                seg= seg + word+' '                   
            texts.append(seg)

            # 每處理完1000篇文章，報告進度。******    
            if i%1000==0:
                print (i, 'done')   
            elif i==len(doc_list):
                print ('finished')         
        else:
            texts.append("")
    else:
        texts.append("")
```

```python
df['token']=texts
```
