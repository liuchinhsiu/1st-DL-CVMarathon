#### 刪除不要的欄位

```python
# app_train1= app_train.drop("Unnamed: 21", axis = 1)
# app_train1

app_train = app_train.drop(columns=['Unnamed: 3', 'Unnamed: 21', 'Unnamed: 22','Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26'])
```

```python
#用sort_values來做資料排序
app_train.sort_values(by = 'age').head(30)
```

#### 截取部分資料

```python
##擷取特定條件資料

fliter = (app_train['age'] > 66)
app_train[fliter]

app_train[app_train['age'] > 66]


#isin()
mask = app_train["country"].isin(["UK","India","Spain"])
app_train[mask]

##找到非空值

mask = app_train["symptom_onset"].notnull()
app_train[mask]

#用兩個比較運算子取出範圍，但其實還有一個方法可以取中間值就是.between()
```
