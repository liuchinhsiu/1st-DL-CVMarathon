```python
# 秀出資料欄位的類型, 與對應的數量
# df.dtypes : 轉成以欄位為 index, 類別(type)為 value 的 DataFrame
# .reset_index() : 預設是將原本的 index 轉成一個新的欄位, 如果不須保留 index, 則通常會寫成 .reset_index(drop=True)
dtype_df = df.dtypes.reset_index() 
dtype_df.columns = ["Count", "Column Type"]
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df
```

```python
# 確定只有 int64, float64, object 三種類型後對欄位名稱執行迴圈, 分別將欄位名稱存於三個 list 中
int_features = []
float_features = []
object_features = []
# .dtypes(欄位類型), .columns(欄位名稱) 是 DataFrame 提供的兩個方法, 這裡順便展示一下 for 與 zip 搭配的用法
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else:
        object_features.append(feature)
# 這邊採用的寫法稱為 f-string, 是 Python 3.6.2 以後版本才出現的
# 如果無法執行, 則需要更新到這個版本之後, 或自行將程式改寫為 str.format 形式
# 改寫方式可以參考 https://blog.louie.lu/2017/08/08/outdate-python-string-format-and-fstring/
print(f'{len(int_features)} Integer Features : {int_features}\n')
print(f'{len(float_features)} Float Features : {float_features}\n')
print(f'{len(object_features)} Object Features : {object_features}')
```

用密度圖和直方圖探討信用卡金額的分布狀態
-> 可以發現數量最多的金額集中在$0 - $1,000,000間

```python
#信用卡金額 -> 用密度圖表示
df['AMT_CREDIT'].plot.kde(bw_method=3, figsize=(8,4))
```

```python
#信用卡金額 -> 用數量表示

df['AMT_CREDIT'].plot.hist(bins=12, alpha=0.5, figsize=(8,4))
```

用密度圖和直方圖探討收入的分布狀態
-> 可以發現數量最多的金額集中在比0.0*1e8 - o.1*1e8 (約為10**7)

```python
#收入 -> 用密度圖表示
df['AMT_INCOME_TOTAL'].plot.kde(bw_method=3, figsize=(8,4))
```

```python
#收入-> 用數量表示

df['AMT_INCOME_TOTAL'].plot.hist(bins=12, alpha=0.5, figsize=(8,4))
```

```python
#探討收入和信用卡費用的關係 -> 沒有正向關係
ax1 = df.plot.scatter(x='AMT_INCOME_TOTAL',
                      y='AMT_CREDIT',
                      c='DarkBlue')
```

探討target = 1 / 0 各自的收入和信用卡金額的分布情形
->target為1, 信用卡費比為0更集中於0-1000000
->target為1, 收入更加集中於0.0 *1e8的位置

```python
df_1 = df[df['TARGET'] == 1]
df_0 = df[df['TARGET'] == 0]
```

```python
df_1['AMT_CREDIT'].plot.kde(bw_method=3, figsize=(8,4))
df_0['AMT_CREDIT'].plot.kde(bw_method=3, figsize=(8,4))
```

```python
df_1['AMT_INCOME_TOTAL'].plot.kde(bw_method=3, figsize=(8,4))
df_0['AMT_INCOME_TOTAL'].plot.kde(bw_method=3, figsize=(8,4))
```
